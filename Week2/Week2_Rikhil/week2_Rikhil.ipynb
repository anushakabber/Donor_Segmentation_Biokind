```python
import math
import pandas as pd
from datetime import datetime
```

```python
# Average Count (Email.Address.Opt.Out.Date) per region
df = pd.read_csv("final_df.csv")
df.head()
```

```python
# Make sure that both of these columns are dattimes so we can then subtract them later
df['Email.Send.Date.Time'] = pd.to_datetime(df['Email.Send.Date.Time'])
df['Email.Address.Opt.Out.Date'] = pd.to_datetime(df['Email.Address.Opt.Out.Date'])

# Create a new column that tells us about the duration between the email sent and the opt out date
df['Out_Out_Duration'] = df['Email.Address.Opt.Out.Date'] - df['Email.Send.Date.Time']
```

```python
grouped_df = df.groupby(['Zip.Code'])
grouped_df.head()
```

```python
# Question 4: How many times on average are people opting out per region?
region_Opt_Out_Counts = grouped_df['Email.Address.Opt.Out.Date'].count()
region_Opt_Out_Counts_mean = region_Opt_Out_Counts.mean()

print("Opt out count average acrosss all regions is: " + str(region_Opt_Out_Counts_mean))
print(region_Opt_Out_Counts.head())
```

```python
# Question 5: How long does it take someone on average to opt out?

# Calculate the mean and then drop any values that are NaT
avg_Opt_Out_Duration = grouped_df['Out_Out_Duration'].mean()
avg_Opt_Out_Duration_Dropped = avg_Opt_Out_Duration.dropna()

print("On average by region, it takes approximately " + str(avg_Opt_Out_Duration_Dropped.mean()) + " for someone to opt out")

avg_Opt_Out_Duration_Dropped.head()
```

```python
# Question 9: Average donors per region

# Filter the donations so that we only count donations above 0 (this should inherently be true but jus tmaking sure)
# Then group by region and find all the unique donors for that region
unique_Donation_Above_Zero = df[df['Largest.Donation.Amount'] > 0].groupby('Zip.Code')['Account.ID'].nunique()

# Calculate the average for all regions
mean_Unique_Donation_Above_Zero = unique_Donation_Above_Zero.mean()
print("The average donors per region is " + str(mean_Unique_Donation_Above_Zero))
unique_Donation_Above_Zero.head()
```

```python
# Convert all the data into CSV files
region_Opt_Out_Counts.to_csv("Question4.csv")
avg_Opt_Out_Duration_Dropped.to_csv("Question5.csv")
unique_Donation_Above_Zero.to_csv("Question9.csv")
```
