```python
import pandas as pd
import math
from datetime import datetime
```

```python
df = pd.read_csv('cleaned_merged_df_ZipCode.csv')
df
```

```python
open_count_zip = df[ df["Zip.Code"] == '53538']["Open.Count"]
tmp = list()
for count in open_count_zip:
    if math.isnan(count):
        tmp.append(0)
    else:
        tmp.append(count)
print(sum(tmp))
```

```python
## 1. Avg. Open.Count per region
## find the unique values of Zip Code column
unique_values = df["Zip.Code"].unique()
## create a dictionary to store zip code and average open rate
count_zip = dict()
## loop through the every zip code
for zip in unique_values.tolist():
    ## create a list for every zip code
    count_zip[zip] = list()
    ## create a data frame for the zip and it's Open.Count column
    open_count_zip = df[ df["Zip.Code"] == zip]["Open.Count"]
    ## loop through open_count_zip 
    for count in open_count_zip:
        ## if count is nan add 0 to list else add count
        if math.isnan(count):
            continue
        else:
            count_zip[zip].append(count)
    ## get the average of open rate of each zip code
    if len(count_zip[zip]) == 0:
        count_zip[zip] = 0
    else:
        count_zip[zip] = sum(count_zip[zip]) / len(count_zip[zip])

print(count_zip)
```

```python
## 2. Avg. email.link clicks per region
## find the unique values of Zip Code column
unique_values = df["Zip.Code"].unique()
## create a dictionary to store zip code and average open rate
click_zip = dict()
## loop through the every zip code
for zip in unique_values.tolist():
    ## create a list for every zip code
    click_zip[zip] = list()
    ## create a data frame for the zip and it's Email.Link.Click.Count column
    click_count_zip = df[ df["Zip.Code"] == zip]["Email.Link.Click.Count"]
    ## loop through open_count_zip 
    for count in click_count_zip:
        ## if count is nan add 0 to list else add count
        if math.isnan(count):
            continue
        else:
            click_zip[zip].append(count)
     ## get the average of open rate of each zip code
    if len(click_zip[zip]) == 0:
        click_zip[zip] = 0
    else:
        click_zip[zip] = sum(click_zip[zip]) / len(click_zip[zip])

print(click_zip)
```

```python
## 3. Avg. Email.Send.Date.Time - Email.First.Open.Date per region
## find the unique values of Zip Code column
unique_values = df["Zip.Code"].unique()
## create a dictionary to store zip code and average open rate
days_zip = dict()
## loop through the every zip code
for zip in unique_values.tolist():
    ## create a list for every zip code
    days_zip[zip] = list()
    ## create a data frame for the zip and Email.Send.Date.Time, Email.First.Open.Date and drop the null value rows
    zip_all = df[ df["Zip.Code"] == zip]
    days_count_zip = zip_all[['Email.Send.Date.Time', 'Email.First.Open.Date']].dropna()
    ## loop through open_count_zip 
    for index, row in days_count_zip.iterrows():
        # Parse 'Email.Send.Date.Time' and 'Email.First.Open.Date' as dates
        try:
            send_date = row['Email.Send.Date.Time']
            open_date = row['Email.First.Open.Date']
            
            send_date = send_date[0:10].split("/")
            month1 = int(send_date[0])
            date1 = int(send_date[1])
            year1 = int(send_date[2])
            open_date = open_date.split("/")
            month2 = int(open_date[0])
            date2 = int(open_date[1])
            year2 = int(open_date[2])

            date1 = datetime(year1, month1, date1)
            date2 = datetime(year2, month2, date2)
            days_zip[zip].append((date2 - date1).days)
        except ValueError:
            continue
    ## get the average of open rate of each zip code
    if len(days_zip[zip]) == 0:
        days_zip[zip] = 0
    else:
        days_zip[zip] = sum(days_zip[zip]) / len(days_zip[zip])


print(days_zip)
```
