```python
import pandas as pd
import math
```

```python
df = pd.read_csv('/Users/alysiachou/Documents/Donor_Segmentation_Biokind/Week3/results_df.csv')
df
```

```python
high_high = df[df['Prioritized Donor Kind'] == 'High Engagement-High Donation']
unique_values = high_high["Zip.Code"].unique()
```

```python
campaign_zip = dict()
for zip in unique_values.tolist():
    campaign_zip[zip] = dict()
    c_zip = high_high[ high_high["Zip.Code"] == zip]["Email.Campaign.Name"]
    for name in c_zip:
        if name in campaign_zip[zip]:
            campaign_zip[zip][name] = campaign_zip[zip][name]+1
        else:
            campaign_zip[zip][name] = 1
campaign_zip
```

```python
df_campaign_zip = pd.DataFrame.from_dict(campaign_zip, orient='index').fillna(0).astype(int)

# Reset index to make Zip Code a column
df_campaign_zip = df_campaign_zip.reset_index().rename(columns={"index": "Zip Code"})

# Save the DataFrame to a CSV file
df_campaign_zip.to_csv('campaign_zip.csv', index=False)
```

```python
## number of people that are engaged or donated in a zip code
person_zip = dict()
for zip in unique_values.tolist():
    person_zip[zip] = list()
    account_zip = high_high[ high_high["Zip.Code"] == zip]["Account.ID"]
    for id in account_zip:
        if id in person_zip[zip]:
            continue
        else:
            person_zip[zip].append(id)
    person_zip[zip] = len(person_zip[zip])
    

print(person_zip)
```

```python
df_person_zip = pd.DataFrame(list(person_zip.items()), columns=['Zip.Code', 'Number of people'])
df_person_zip.to_csv('person_zip.csv', index=False)
```
