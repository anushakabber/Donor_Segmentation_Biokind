```python
import pandas as pd
```

```python
df_result = pd.read_csv("/Users/alysiachou/Documents/Donor_Segmentation_Biokind/Week3/results_df.csv")
df_metro_income = pd.read_csv("/Users/alysiachou/Documents/Donor_Segmentation_Biokind/Week4/metro_income_df.csv")
df_metro_income

```

```python
## change the value type to int
df_metro_income["PerCapitaPersonalIncome"] = df_metro_income["PerCapitaPersonalIncome"].str.replace(",", "", regex=True)

df_metro_income['PerCapitaPersonalIncome'] = pd.to_numeric(df_metro_income["PerCapitaPersonalIncome"], errors="coerce")
```

```python
## get the dataframe with the High Donation
df_high_donation_result = df_result[ df_result["Donation Type"] == "High Donation"]
## get unique zip codes
unique_zip_high = df_high_donation_result["Zip.Code"].unique()
```

```python
## calculate the mean income from the zip codes from above
high_donation_income = df_metro_income[df_metro_income["Zip Code"].isin(unique_zip_high)]

high_donation_income = high_donation_income.dropna(subset=["PerCapitaPersonalIncome"])

mean_income = high_donation_income["PerCapitaPersonalIncome"].mean()
```

```python
## set upper and lower bound
lower_bound = mean_income - 5000  
upper_bound = mean_income + 5000

## get the zip codes that are not in result_df and the income is within the bounds
similar_income_zips = df_metro_income[
    (df_metro_income["PerCapitaPersonalIncome"] >= lower_bound) &
    (df_metro_income["PerCapitaPersonalIncome"] <= upper_bound) &
    (~df_metro_income["Zip Code"].isin(unique_zip_high))
]
df_final = similar_income_zips[ ["Zip Code", "PerCapitaPersonalIncome"]]
df_final
```

```python
df_final.to_csv('zip_percapitapersonalincome.csv', index=False)
```
