```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/armaanjain/Documents/UW Freshman (2024-2025)/Clubs/BioKind Analytics/Donor_Segmentation_Biokind-main/Week4/final_metro_income_df.csv')

# One-hot encode columns Donation Type and Engagement Type
encoded_df = pd.get_dummies(df, columns=['Donation Type', 'Engagement Type'])

# Select columns to include in the correlation matrix
selected_columns = ['Population', 'Density', 'PerCapitaPersonalIncome', 'RUCC_2023'] + [col for col in encoded_df.columns if col.startswith('Donation Type_') or col.startswith('Engagement Type_')]

# Calculate the correlation matrix for selected columns
correlation_matrix = encoded_df[selected_columns].corr()

# Plot the correlation matrix using Seaborn
plt.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix")
plt.show()
```
