```python
import pandas as pd
```

```python
merged_df = pd.read_csv('merged_data.csv')
```

```python
merged_df.info()
```

```python
merged_df['Largest.Donation.Amount'] = merged_df['Largest.Donation.Amount'].apply(lambda x: x.replace(',', ''))
```

```python
merged_df = merged_df.astype({'Largest.Donation.Amount': 'float64'})
```

```python
merged_df.drop(columns=['X.y', 'X.x', 'Invalid.Email.Address', 'Email.Final.Receipt', 'Salutation'], inplace=True)
```

```python
for col in merged_df.columns: 
    if 'date' in col.lower(): 
        merged_df[col] = pd.to_datetime(merged_df[col])
```

```python
merged_df.info()
```

```python
merged_df.rename(columns={'Email.Recipient.Full.Name..F.': 'Email.Recipient.Full.Name', 'Address.Line.1': 'Address', 'Full.Name..F.': 'Full.Name'}, inplace=True)
```

```python
merged_df.to_csv('cleaned_merged_df.csv')
```

```python
def compare_columns(df, col1, col2):
    unique_in_col1 = merged_df[col1].isin(merged_df[col2])
    unique_in_col2 = merged_df[col2].isin(merged_df[col1])
    return df[col1][~unique_in_col1], df[col2][~unique_in_col2]

# Remove leading and trailing spaces to help compare the 2 columns better
merged_df['Email.Recipient.Full.Name'] = merged_df['Email.Recipient.Full.Name'].str.strip()
merged_df['Full.Name'] = merged_df['Full.Name'].str.strip()

# Standardizes amount of space between names to help compare the 2 columns better
merged_df['Email.Recipient.Full.Name'] = merged_df['Email.Recipient.Full.Name'].str.split().str.join(' ')
merged_df['Full.Name'] = merged_df['Full.Name'].str.split().str.join(' ')

unique_in_A, unique_in_B = compare_columns(merged_df, 'Email.Recipient.Full.Name', 'Full.Name')

# it appears that Full.Name and Email.Recipient.Full.Name have many unique values but a lot of them reference the same name just different spellings and nicknames
print("Unique values in A:", unique_in_A.values)
print("Unique values in B:", unique_in_B.values)
```
