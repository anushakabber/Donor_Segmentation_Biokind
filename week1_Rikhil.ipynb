```python
import pandas as pd
```

```python
merged_df = pd.read_csv('merged_data.csv')
```

```python
merged_df['Largest.Donation.Amount'] = merged_df['Largest.Donation.Amount'].apply(lambda x: x.replace(',', ''))
```

```python
merged_df = merged_df.astype({'Largest.Donation.Amount': 'float64'})
```

```python
merged_df.drop(columns=['X.y', 'X.x', 'Invalid.Email.Address', 'Email.Final.Receipt', 'Salutation'], inplace=True)
```

```python
for col in merged_df.columns: 
    if 'date' in col.lower(): 
        merged_df[col] = pd.to_datetime(merged_df[col])
```

```python
merged_df.info()
```

```python
merged_df.rename(columns={'Email.Recipient.Full.Name..F.': 'Email.Recipient.Full.Name', 'Address.Line.1': 'Address', 'Full.Name..F.': 'Full.Name'}, inplace=True)
```

```python
# Super scuffed way to fix all the city data, there is a probably a nicer way but this gets the job done
def fixSegmentCasing(word):
    if (type(word) is float):
        return word
    
    lowerCaseCity = word.strip().lower()
    useSpace = False

    if (lowerCaseCity[-4:] == ", wi"):
        lowerCaseCity = lowerCaseCity[:-4]
    elif (lowerCaseCity[-3:] == " wi"):
        lowerCaseCity = lowerCaseCity[:-3]

    if (lowerCaseCity[:2].lower() == "mc"):
        lowerCaseCity = lowerCaseCity.replace(" ", "")
        words = [lowerCaseCity[:2], lowerCaseCity[2:]]
        useSpace = False
    else:
        words = lowerCaseCity.split(" ")
        useSpace = True

    fixedCasing = ""
    for w in words:
        if useSpace:
            fixedCasing += w[0].upper() + w[1:] + " "
        else:
            fixedCasing += w[0].upper() + w[1:]

    return fixedCasing.strip()

merged_df['City'] = merged_df['City'].apply(fixSegmentCasing)
```
